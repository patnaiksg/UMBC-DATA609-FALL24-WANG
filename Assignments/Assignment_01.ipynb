## Step 2 - Explore Student Age

```python
yob_list = list(df["8. My Birth Year"])
yob_list
```

```python
cleansed_yob_list = []
cleansed_yob_list
```

```python
for yob in yob_list:
  cleansed_yob = yob[-4:]
  cleansed_yob_list.append(cleansed_yob)

cleansed_yob_list
```

```python
age_list = []

for yob in cleansed_yob_list:
  age = 2024 - int(yob)
  age_list.append(age)

age_list
```

```python
min(age_list)
```

```python
max(age_list)
```

```python
total_age = 0

for age in age_list:
  total_age = total_age + age

total_age
```

```python
average_age = total_age / len(age_list)

round(average_age)
```

```python
total_age = 0

for age in age_list:
  if age>=35 or age<=22:
    print("Age", age, "is skipped")
  else:
    total_age = total_age + age

print("Total age is", total_age)

average = round(total_age / (len(age_list)))

print("Average age is", average)
```

```python
total_age = 0

for age in age_list:
  if age>=35 or age<=22:
    print("Age", age, "is skipped")
    #continue

  total_age = total_age + age

print("Total age is", total_age)

average = round(total_age / (len(age_list)))

print("Average age is", average)
```
