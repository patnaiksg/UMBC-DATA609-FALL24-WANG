Pick a few countries of your interest (for example, G7, BRICS, Nordic countries).
- 1 Create a line chart showing the trend of Life Expectancy over time for comparing those countries.
- 2 Pick a year from the data of the above few countries and create a scatter plot to compare their GDP per capita and life expectancy. The size of the dots represent the population. This type of scatter plot is called bubble chart.
- 3 Bonus - compute the average GDP per capita, the average life expectancy, and the total population for each year of all the seven regions (aggregation) and repeat 1 and 2 using the aggregated data for all regions

### 1. Create a line chart showing the trend of Life Expectancy over time for comparing those countries.

```python
g7 = ['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States']

df_g7 = df[df['Country'].isin(g7)].dropna(subset=['GDP per capita (current US$)'])

df_g7.sample(5)
```

```python
import plotly.express as px

fig = px.line(df_g7,
              x="Year",
              y="Life expectancy at birth, total (years)",
              color="Country",
              title="Life Expectancy Trends Over Time for G7 Countries",
              labels={"Life expectancy at birth, total (years)": "Life Expectancy (years)"})

fig.show()
```

### 2. Pick a year from the data of the above few countries and create a scatter plot to compare their GDP per capita and life expectancy.

```python
# Pick a year from the data of the above few countries and create a scatter plot to compare their GDP per capita and life expectancy.
year_of_interest = 2018

df_g7_2018 = df_g7[df_g7["Year"] == year_of_interest]

# The size of the dots represent the population. This type of scatter plot is called bubble chart.
fig = px.scatter(
    df_g7_2018,
    x="GDP per capita (current US$)",
    y="Life expectancy at birth, total (years)",
    color="Country",
    text="Country",
    size="Population, total",
    title=f"Relationship between GDP per Capita and Life Expectancy in {year_of_interest}",
    labels={
        "GDP per capita (current US$)": "GDP per Capita (Current US$)",
        "Life expectancy at birth, total (years)": "Life Expectancy (Years)"
    },
    height=600,
    width=1000
)

# Update text position
fig.update_traces(textposition='top center')
fig.update_layout(showlegend=False)

fig.show()
```

### 3. Bonus - compute the average GDP per capita, the average life expectancy, and the total population for each year of all the seven regions (aggregation) and repeat 1 and 2 using the aggregated data for all regions

```python
# 3. Bonus - compute the average GDP per capita, the average life expectancy, and the total population for each year of all the seven regions (aggregation)
# and repeat 1 and 2 using the aggregated data for all regions

df_g7_aggregated = df_g7.groupby(['Year', 'Region']).agg(
    avg_gdp_per_capita=('GDP per capita (current US$)', 'mean'),
    avg_life_expectancy=('Life expectancy at birth, total (years)', 'mean'),
    total_population=('Population, total', 'sum')
).reset_index()

df_g7_aggregated.sample(5)
```

```python
fig = px.line(df_g7_aggregated,
              x="Year",
              y="avg_life_expectancy",
              color="Region",
              title="Average Life Expectancy Over Time by Region",
              labels={"avg_life_expectancy": "Average Life Expectancy (years)"})

fig.show()
```

```python
fig2 = px.line(df_g7_aggregated,
               x="Year",
               y="avg_gdp_per_capita",
               color="Region",
               title="Average GDP per Capita Over Time by Region",
               labels={"avg_gdp_per_capita": "Average GDP per Capita (US$)"})

fig2.show()
```

## Exercise

Calculate average GDP per capita for  each region and make a bar chart.

```python

df_cleaned = df[['Region', 'GDP per capita (current US$)']].dropna()

average_gdp_by_region = df_cleaned.groupby('Region').agg({'GDP per capita (current US$)': 'mean'}).reset_index()
fig = px.bar(
    data_frame=average_gdp_by_region,
    x='Region',
    y='GDP per capita (current US$)',
    title='Average GDP per Capita by Region',
    labels={
        'GDP per capita (current US$)': 'Average GDP per Capita (Current US$)',
        'Region': 'Region'
    },
    text='GDP per capita (current US$)'
)


fig.show()
```
