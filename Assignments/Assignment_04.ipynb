## Homework Question 1:

to check if this dataset has duplicate rows. If it has, only keep one from the duplicated rows and drop the rest.

```python
duplicates = df3.duplicated()

# Count the number of duplicate rows
num_duplicates = duplicates.sum()
print(f"Number of duplicate rows: {num_duplicates}")
```

```python
if num_duplicates > 0:
    df3 = df3.drop_duplicates()

# Display the number of duplicates found and confirm the new shape of the dataset
print(num_duplicates)
```

## HW Question 2

Find out if there is a relationship between wealth and health

Pick any year worth of data and use plotly express to generate a scatter plot with GDP as x and life expectancy as y.

```python
df_2019 = df3[df3['Year'] == 2019]

# Generate a scatter plot with GDP per capita as x and life expectancy as y using plotly express
fig = px.scatter(df_2019,
                 x="GDP per capita (current US$)",
                 y="Life expectancy at birth, total (years)",
                 title="Relationship Between Wealth (GDP per capita) and Health (Life Expectancy) in 2020",
                 labels={"GDP per capita (current US$)": "GDP per Capita (US$)",
                         "Life expectancy at birth, total (years)": "Life Expectancy (Years)"},
                 trendline="ols")

# Show the plot
fig.show()
```

From the analysis, we can see that with increasing GDP per Capita, the life expectancy is also increasing with a life expectancy of 75-90 years in 2019 in  countries with higher GDP
