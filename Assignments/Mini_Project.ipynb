## Step 4 - Categorize Each Country

Create a new column "Group". If a country is one of the seven G7 countried, the value of "Group" would be "G7", if a country of one of the five BRICS countries, the value of "Group" would be "BRICS". Otherwise, the value of "Group" would be "REST of WORLD".

```python
# Defining G7 and BRICS countries
g7_countries = ['Canada', 'France', 'Germany', 'Italy', 'Japan', 'United Kingdom', 'United States']
brics_countries = ['Brazil', 'Russia', 'India', 'China', 'South Africa']

# Creating the "Group" column based on country classification
df['Group'] = df['Country Name'].apply(
    lambda x: 'G7' if x in g7_countries else ('BRICS' if x in brics_countries else 'REST OF WORLD')
)

# Display the first few rows to verify the new column
df[['Country Name', 'Group']].sample(5)
```

## Step 5 - Aggregate Data

Aggregate the following columns at the "Group" level for each year:

- For GDP, using "sum" for aggregation.

```python
# Aggregating the GDP column at the "Group" level for each year using the sum
gdp_aggregated = df.groupby(['Year', 'Group'])['GDP ($Billions)'].sum().reset_index()

# Display the aggregated results
gdp_aggregated.tail()
```

- For GDP Per Capita, use "mean" for aggregation.

```python
# Aggregating the GDP Per Capita column at the "Group" level for each year using the mean
gdp_per_capita_aggregated = df.groupby(['Year', 'Group'])['GDP Per Capita ($)'].mean().reset_index()

# Display the aggregated results
gdp_per_capita_aggregated.head()
```

- Checked calculation with the reference site.

## Step 6 - Calculate Percentage Share of GDP

For each year, calculate the percentage share of each group's GDP. Check your calculation with the reference site.

```python
total_gdp_per_year = df.groupby('Year')['GDP ($Billions)'].sum().reset_index(name='Total GDP')

df_with_total_gdp = df.merge(total_gdp_per_year, on='Year')

df_with_total_gdp['GDP Share (%)'] = (
    df_with_total_gdp['GDP ($Billions)'] / df_with_total_gdp['Total GDP'] * 100
)

gdp_share_aggregated = df_with_total_gdp.groupby(['Year', 'Group'])['GDP Share (%)'].sum().reset_index()

gdp_share_aggregated.head()
```

## Step 7 - Plot the GDP Percentage Share Over Time

```python
import plotly.express as px

fig = px.line(
    gdp_share_aggregated,
    x="Year",
    y="GDP Share (%)",
    color="Group",
    title="GDP Percentage Share Over Time",
    labels={"GDP Share (%)": "GDP Share (%)", "Year": "Year", "Group": "Group"}
)

fig.update_layout(
    xaxis=dict(title="Year"),
    yaxis=dict(title="GDP Share (%)"),
    legend=dict(title="Group")
)

fig.show()
```

## Step 8 - Plot the Total GDP Over Time

```python
total_gdp_per_year = df.groupby('Year')['GDP ($Billions)'].sum().reset_index(name='Total GDP')

fig = px.line(
    total_gdp_per_year,
    x="Year",
    y="Total GDP",
    title="Total GDP Over Time",
    labels={"Total GDP": "Total GDP (Billions)", "Year": "Year"}
)

fig.update_layout(
    xaxis=dict(title="Year"),
    yaxis=dict(title="Total GDP (Billions)")
)

fig.show()
```

## Step 9 - Plot the 2023 GDP Pie Chart

Each group is a slice of a pie. Make sure the following information is displayed for each slice:

- Group Name
- GDP Percentage Share
- GDP Total

```python
print(df['Year'].unique())
```

```python
import plotly.express as px

# Filter the dataset for the year 2023 and calculate the GDP percentage share for each group
gdp_2023 = df[df['Year'] == 2023]
total_gdp_2023 = gdp_2023['GDP ($Billions)'].sum()

# Add percentage share column
gdp_2023['GDP Share (%)'] = (gdp_2023['GDP ($Billions)'] / total_gdp_2023) * 100

# Aggregate the data at the group level
gdp_2023_grouped = gdp_2023.groupby('Group').agg(
    Total_GDP=('GDP ($Billions)', 'sum'),
    GDP_Share=('GDP Share (%)', 'sum')
).reset_index()

fig = px.pie(
    gdp_2023_grouped,
    values="Total_GDP",
    names="Group",
    title="GDP Distribution by Group (2023)",
    hover_data=["GDP_Share", "Total_GDP"],
    labels={"GDP_Share": "GDP Share (%)", "Total_GDP": "Total GDP (Billions)"}
)

fig.update_traces(
    textposition='inside',
    textinfo='label+percent'
)

fig.show()
```

## Step 10 - Plot the 2023 GDP Pie Charts

One pie chart for G7 and one pie chart for BRICK. Each country is a slice of a pie. Make sure the following information is displayed for each slice:

- Country Name
- GDP Percentage Share
- GDP Total

```python
import plotly.express as px

# Filter data for the year 2023 and ensure it is a copy
gdp_2023 = df[df['Year'] == 2023].copy()

# Calculate total GDP for 2023 for each group
total_gdp_2023_g7 = gdp_2023[gdp_2023['Group'] == 'G7']['GDP ($Billions)'].sum()
total_gdp_2023_brics = gdp_2023[gdp_2023['Group'] == 'BRICS']['GDP ($Billions)'].sum()

# Add GDP percentage share column for G7
gdp_2023.loc[gdp_2023['Group'] == 'G7', 'GDP Share (%)'] = (
    gdp_2023['GDP ($Billions)'] / total_gdp_2023_g7 * 100
)

# Add GDP percentage share column for BRICS
gdp_2023.loc[gdp_2023['Group'] == 'BRICS', 'GDP Share (%)'] = (
    gdp_2023['GDP ($Billions)'] / total_gdp_2023_brics * 100
)

# Create G7 pie chart
g7_data = gdp_2023[gdp_2023['Group'] == 'G7']
fig_g7 = px.pie(
    g7_data,
    values="GDP ($Billions)",
    names="Country Name",
    title="GDP Distribution for G7 Countries (2023)",
    hover_data=["GDP Share (%)", "GDP ($Billions)"],
    labels={"GDP Share (%)": "GDP Share (%)", "GDP ($Billions)": "GDP Total (Billions)"}
)

fig_g7.update_traces(
    textposition='inside',
    textinfo='label+percent'
)

# Create BRICS pie chart
brics_data = gdp_2023[gdp_2023['Group'] == 'BRICS']
fig_brics = px.pie(
    brics_data,
    values="GDP ($Billions)",
    names="Country Name",
    title="GDP Distribution for BRICS Countries (2023)",
    hover_data=["GDP Share (%)", "GDP ($Billions)"],
    labels={"GDP Share (%)": "GDP Share (%)", "GDP ($Billions)": "GDP Total (Billions)"}
)

fig_brics.update_traces(
    textposition='inside',
    textinfo='label+percent'
)

# Show the plots
fig_g7.show()
fig_brics.show()
```

## Step 14 - Plot the 2023 GDP Per Capita Bar Chart

Each group is a bar.

```python
# Calculate the mean GDP per capita for each group in 2023
gdp_per_capita_2023 = df[df['Year'] == 2023].copy()
gdp_per_capita_by_group = gdp_per_capita_2023.groupby('Group')['GDP Per Capita ($)'].mean().reset_index()

# Create a bar chart
fig = px.bar(
    gdp_per_capita_by_group,
    x="Group",
    y="GDP Per Capita ($)",
    title="2023 GDP Per Capita by Group",
    labels={
        "Group": "Group",
        "GDP Per Capita ($)": "GDP Per Capita ($)"
    },
    text="GDP Per Capita ($)"  # Display values on bars
)

# Customize layout for better readability
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.update_layout(
    xaxis=dict(title="Group"),
    yaxis=dict(title="GDP Per Capita ($)"),
    showlegend=False
)

# Show the plot
fig.show()
```

## Step 15 - Plot the 2023 GDP Per Capita Bar Chart of G7 and BRICS countries

Each country is a bar. Differentiate group by color.

```python
# Filter the data for the year 2023 and ensure only G7 and BRICS countries are included
gdp_per_capita_2023 = df[(df['Year'] == 2023) & (df['Group'].isin(['G7', 'BRICS']))].copy()

# Create a bar chart
fig = px.bar(
    gdp_per_capita_2023,
    x="Country Name",
    y="GDP Per Capita ($)",
    color="Group",
    title="2023 GDP Per Capita: G7 and BRICS Countries",
    labels={
        "Country Name": "Country",
        "GDP Per Capita ($)": "GDP Per Capita ($)",
        "Group": "Group"
    },
    text="GDP Per Capita ($)"
)

# Customize layout for better readability
fig.update_traces(texttemplate='%{text:.2f}', textposition='outside')
fig.update_layout(
    xaxis=dict(title="Country"),
    yaxis=dict(title="GDP Per Capita ($)"),
    legend=dict(title="Group"),
    barmode="group"
)

# Show the plot
fig.show()
```

## Step 16 - Create a Choropleth

To show G7 countries and BRICS countries and use color code to differentiate the two broups.

```python
# Filter the data to include only G7 and BRICS countries
g7_brics = df[df['Group'].isin(['G7', 'BRICS'])].copy()

# Create a choropleth map
fig = px.choropleth(
    g7_brics,
    locations="Country Code",  # ISO country codes
    color="Group",  # Color by group (G7 or BRICS)
    hover_name="Country Name",  # Show country name on hover
    title="Choropleth Map: G7 and BRICS Countries",
    labels={"Group": "Group"}
)

# Customize the layout for better readability
fig.update_geos(
    showcoastlines=True,
    coastlinecolor="LightGray",
    showland=True,
    landcolor="whitesmoke",
    showocean=True,
    oceancolor="lightblue"
)

# Show the plot
fig.show()
```
